```python
# Scatter Plot of Race/Ethnicity vs. Average Score

import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
url = '/content/gdrive/MyDrive/cleaned_data (1).csv'
data = pd.read_csv(url)
# Assuming your dataset has columns 'gender' and 'Average Score'
race = data['race/ethnicity']
average_scores = data['average score']

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(race, average_scores, alpha=0.5)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')
plt.title('Scatter Plot of Race/Ethnicity vs. Average Score')

# Save the histogram
plt.savefig('/content/gdrive/MyDrive/scatter_plot.png', bbox_inches='tight')
# Show the histogram
plt.show()
```
